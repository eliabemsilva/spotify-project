# Transformando arquivo JSON em CSV

### Importando a biblioteca Pandas e lendo os arquivos JSON

```python
import pandas as pd

dados_22_23 = pd.read_json('StreamingHistory0.json')
display(dados_22_23)

dados_23_24 = pd.read_json('StreamingHistory_music_0.json')
display(dados_23_24)
```

### Filtrando o data frame com base na coluna 'endTime'

```python
dados_23_24 = dados_23_24[dados_23_24['endTime'] >= '2023-04-22']
dados_23_24
```

### Unificando os 2 data frames

```python
dados = pd.concat([dados_22_23, dados_23_24], ignore_index=True)
dados
```

### Transformando a coluna 'msPlayed' de milissegundos em minutos

```python
dados['msPlayed'] = dados['msPlayed'] / 60000
dados
```

### Renomeando colunas para português

```python
dados.rename(columns={'endTime': 'dataTermino', 'artistName': 'nomeArtista', 'trackName': 'nomeFaixa', 'msPlayed': 'minutosTocados'}, inplace = True)
dados
```

### Extraindo somente a data da coluna 'dataTermino'

```python
dados['dataTermino'] = dados['dataTermino'].str.split().str[0]
dados
```

   ### Concatenando colunas 'nomeFaixa' e 'nomeArtista' como chave

```python
dados['idFaixaArtista'] = dados['nomeFaixa'].str.cat(dados['nomeArtista'], sep = '-')
dados
```

### Filtrando linhas que não estão zeradas na coluna 'minutosTocados'

```python
dados = dados[dados['minutosTocados'] != 0]
dados
```

### Removendo colunas 'nomeFaixa' e 'nomeArtista'

```python
dados = dados.drop(columns=['nomeFaixa', 'nomeArtista'])
dados
```

### Gerando arquivo CSV

```python
dados.to_csv('Historico.csv', index=False)
```
